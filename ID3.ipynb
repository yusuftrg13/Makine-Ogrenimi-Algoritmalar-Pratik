```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
import matplotlib.pyplot as plt

```

```python
#Veri Yükleme
data=pd.read_csv('iklim.csv')
```

```python
#LabelEncoder Nesnesi
le=LabelEncoder()

#Tüm sütunları dönüştürelim
for col in data.columns:
  data[col]=le.fit_transform(data[col])

print(data.head())
```

```python
#Özellikler (X) ve Hedef (y)  Değişkenlerini Seçelim
X=data.drop('Oynanabilir',axis=1)

#Hedef y oynanabilir
y=data['Oynanabilir']
```

```python
#Karar Agacı Modeli Kurma
model=DecisionTreeClassifier(criterion='entropy')
model.fit(X,y)
```

```python
#Grafik Halinde Ağacı Görselleştirme
plt.figure(figsize=(12,8))
plot_tree(model,filled=True,feature_names=X.columns,class_names=['Evet','Hayır'])
plt.title('Karar Agacı(Entropi İle)')
```
